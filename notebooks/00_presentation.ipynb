## 1. Project Introduction & Background

### Introduction: In Search of Lost River Memories

Rivers are the cradle of civilization. However, their flow is not eternal.
Throughout the eons, rivers meander and drastically change their form. Water may have long since vanished from places where civilizations once flourished.

But what if those memories have been passed down to the present day as "place names"?

Our adventure began with such a romantic hypothesis. Land names and folklore might be time capsules of the soul that remain rooted in places even after people have left and rivers have dried up.

The name of this project, "Tamagawa to Z," derives from Japan's "Tamagawa River" where our team members live. This river, which once repeatedly flooded, left the same place name "Todoroki" on both its banks. The place names still connect the two lands that the river divided and time separated. Fascinated by the storytelling power of these place names, we decided to embark on a journey to search for the same miracle in the distant Amazon lands.

### Our Compass: LLM and Multilingual Toponymy

Amazon's history is a crossroads of diverse languages. The languages of indigenous peoples and immigrants have mixed together, changed over time, and hidden the true nature of the land behind a complex veil.

Our greatest weapon is the **power of LLM** that breaks through this language barrier and the knowledge of **toponymy (the study of place names)**.

We don't simply search for words meaning "water." We had LLMs interpret countless place names scattered throughout the Amazon in multilingual contexts, comprehensively judging their etymology, semantic similarity, and historical background, practicing what could be called "linguistic taxonomy." This is the work of excavating the "land's memory" sleeping deep within words.

Through this process, we created the world's one and only **"Multilingual Toponym Dictionary."** This is our unique compass in our long and arduous exploration journey.

### Route to Discovery

We chose **Acre State, Brazil**, filled with possibilities for undiscovered findings, as the stage for our adventure. This land, where "new ground surfaces" are continuously appearing due to deforestation and development, is truly a frontier where numerous geoglyphs have been recently found from satellite images.

Our route is as follows:

**1. Listening to Voices Sleeping in Maps:** With our created "Multilingual Toponym Dictionary" in hand, we picked up place names from Acre State in OSM (OpenStreetMap) and listed places that might be related to water.

**2. Setting Course to the Past:** Our goal is to explore "lost rivers." Therefore, we narrowed down to lands that are a certain distance away from current rivers and have "low wetland levels," suggesting they were possible residential areas in the past. To improve the accuracy of this narrowing down, we adjusted parameters while discussing with LLM (O3 Pro) itself, using how well we could reproduce known ruins as an indicator. It was like experiencing a discussion about unknown sea charts with an AI, an experienced and seasoned navigator.

**3. Relying on Stars:** Eight selected candidate sites. We scored them from multiple perspectives: feature matching combining topography and toponymy at known ruins, possibility of destruction/burial by artifacts, unexplored survey areas, topographical advantages from satellite images, mythological folklore, and more.

### 🎯 Filtering Strategy

To identify "places where water-meaning place names remain despite no current rivers," we set the following conditions:
- **Distance filter**: Points more than 2.5km away from current river channels
- **Water frequency filter**: Points with water occurrence frequency of 15% or less

````python
# Script Execution: Candidate Site Search
print("🔍 Executing ancient river channel candidate site search...")
print("📍 Target area: Acre State")
print("🗂️  Script used: run_site_identification.py")
print("=" * 50)

try:
    # Actual execution command (uses dump file due to long execution time)
    # result = subprocess.run([
    #     "python", "../scripts/run_site_identification.py", 
    #     "--region", "acre",
    #     "--dist-threshold", "2.5",
    #     "--occ-threshold", "15.0",
    #     "--visualize"
    # ], capture_output=True, text=True, timeout=2400)
    
    # Load dump file
    candidates_file = "../data/output/candidates/acre_candidates.csv"
    
    if os.path.exists(candidates_file):
        candidates_data = pd.read_csv(candidates_file)
        print("✅ Loaded candidate site data from dump file")
        print(f"📄 File path: {candidates_file}")
        
        # Data preprocessing (extract coordinate information from geometry column)
        if 'geometry' in candidates_data.columns:
            # Extract coordinates from format POINT (-67.69918823242188 -9.728560447692871)
            def extract_coordinates(geom_str):
                import re
                match = re.search(r'POINT \(([^)]+)\)', str(geom_str))
                if match:
                    coords = match.group(1).split()
                    return float(coords[0]), float(coords[1])  # lon, lat
                return None, None
            
            candidates_data[['lon', 'lat']] = candidates_data['geometry'].apply(
                lambda x: pd.Series(extract_coordinates(x))
            )
        
        # Column name mapping (adapt to actual data)
        if 'type' in candidates_data.columns:
            candidates_data['water_type'] = candidates_data['type']
            
    else:
        # Sample data
        print("⚠️  Dump file not found. Using sample data.")
        print(f"❌ Expected file: {candidates_file}")
        candidates_data = pd.DataFrame({
            'name': ['Olho D\'Água', 'Rio Seco', 'Igarapé Morto', 'Lagoa Seca', 
                    'Cachoeira Velha', 'Poço Antigo', 'Córrego Perdido', 'Água Fria'],
            'lat': [-9.83, -9.45, -9.67, -9.78, -9.52, -9.89, -9.41, -9.73],
            'lon': [-68.50, -68.32, -68.61, -68.43, -68.58, -68.37, -68.49, -68.55],
            'dist_km': [3.2, 4.1, 2.8, 3.7, 5.2, 2.9, 3.4, 4.8],
            'occ_pct': [8.3, 12.1, 5.7, 9.4, 14.2, 6.8, 10.5, 7.9],
            'water_type': ['spring', 'river', 'stream', 'lake', 'waterfall', 'well', 'stream', 'spring']
        })
    
    print(f"📊 Number of extracted candidate sites: {len(candidates_data)}")
    print(f"🗺️  Average river distance: {candidates_data['dist_km'].mean():.1f}km")
    print(f"💧 Average water frequency: {candidates_data['occ_pct'].mean():.1f}%")
    
    # Display top candidate sites (sorted by distance)
    print("\n🎯 Main candidate sites (Top 5 by distance):")
    if len(candidates_data) >= 5:
        top_candidates = candidates_data.nlargest(5, 'dist_km')[['name', 'dist_km', 'occ_pct']]
    else:
        top_candidates = candidates_data[['name', 'dist_km', 'occ_pct']]
    display(top_candidates)
    
    # Detailed data display
    print(f"\n📋 Data columns: {list(candidates_data.columns)}")
    print(f"🔍 Candidate site type distribution:")
    if 'water_type' in candidates_data.columns:
        type_counts = candidates_data['water_type'].value_counts()
        for water_type, count in type_counts.items():
            print(f"   {water_type}: {count} sites")
    elif 'type' in candidates_data.columns:
        type_counts = candidates_data['type'].value_counts()
        for water_type, count in type_counts.items():
            print(f"   {water_type}: {count} sites")
    
except Exception as e:
    print(f"❌ An error occurred: {e}")
    print("📝 In actual operation, pre-executed dump files are used.")
````

````python
# Visualization of Candidate Site Search Results
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Analysis of Candidate Site Search Results', fontsize=18, fontweight='bold')

# 1. Distance distribution
ax1.hist(candidates_data['dist_km'], bins=15, color='lightblue', alpha=0.7, edgecolor='navy')
ax1.axvline(x=2.5, color='red', linestyle='--', linewidth=2, label='Filter threshold (2.5km)')
ax1.set_xlabel('Distance from current river channels (km)')
ax1.set_ylabel('Number of candidate sites')
ax1.set_title('Distance Distribution from Current River Channels', fontweight='bold')
ax1.legend()
ax1.grid(True, alpha=0.3)

# 2. Water frequency distribution
ax2.hist(candidates_data['occ_pct'], bins=15, color='lightgreen', alpha=0.7, edgecolor='darkgreen')
ax2.axvline(x=15.0, color='red', linestyle='--', linewidth=2, label='Filter threshold (15%)')
ax2.set_xlabel('Water occurrence frequency (%)')
ax2.set_ylabel('Number of candidate sites')
ax2.set_title('Water Occurrence Frequency Distribution', fontweight='bold')
ax2.legend()
ax2.grid(True, alpha=0.3)

# 3. Scatter plot (distance vs water frequency)
scatter = ax3.scatter(candidates_data['dist_km'], candidates_data['occ_pct'], 
                     c=range(len(candidates_data)), cmap='viridis', s=100, alpha=0.7)
ax3.axhline(y=15.0, color='red', linestyle='--', alpha=0.5, label='Water frequency threshold')
ax3.axvline(x=2.5, color='red', linestyle='--', alpha=0.5, label='Distance threshold')
ax3.set_xlabel('Distance from current river channels (km)')
ax3.set_ylabel('Water occurrence frequency (%)')
ax3.set_title('Candidate Site Distribution (Distance vs Water Frequency)', fontweight='bold')
ax3.legend()
ax3.grid(True, alpha=0.3)

# 4. Distribution by water system type
# Get types according to actual data
if 'water_type' in candidates_data.columns:
    type_column = 'water_type'
elif 'type' in candidates_data.columns:
    type_column = 'type'
else:
    # Fallback: sample data
    candidates_data['type'] = ['ramal', 'travessa', 'rua', 'other'] * (len(candidates_data) // 4 + 1)
    candidates_data['type'] = candidates_data['type'][:len(candidates_data)]
    type_column = 'type'

type_counts = candidates_data[type_column].value_counts()
bars4 = ax4.bar(range(len(type_counts)), type_counts.values, 
                color=plt.cm.Set3(np.linspace(0, 1, len(type_counts))))
ax4.set_xticks(range(len(type_counts)))
ax4.set_xticklabels(type_counts.index, rotation=45, ha='right')
ax4.set_ylabel('Number of candidate sites')
ax4.set_title('Number of Candidate Sites by Water System Type', fontweight='bold')

# Display numbers on each bar
for bar, count in zip(bars4, type_counts.values):
    height = bar.get_height()
    ax4.text(bar.get_x() + bar.get_width()/2., height + 0.1,
             str(count), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# Analysis of filtering effects
print("\n" + "=" * 60)
print("🔍 Analysis of Filtering Effects")
print("=" * 60)
filter_distance = candidates_data['dist_km'] >= 2.5
filter_occurrence = candidates_data['occ_pct'] <= 15.0
both_filters = filter_distance & filter_occurrence

print(f"🚫 Distance filter (≥2.5km): {filter_distance.sum()} sites passed")
print(f"🚫 Water frequency filter (≤15%): {filter_occurrence.sum()} sites passed")
print(f"✅ Passed both filters: {both_filters.sum()} sites")

if len(candidates_data) > 0:
    max_dist_idx = candidates_data['dist_km'].idxmax()
    print(f"📍 Farthest candidate: {candidates_data.loc[max_dist_idx, 'name']} (distance: {candidates_data.loc[max_dist_idx, 'dist_km']:.1f}km)")

print("=" * 60)
````

## 4. Parameter Optimization

This section uses Bayesian Optimization to optimize analysis parameters and maximize the reproducibility of known ruins.

### 📋 Script Used: `run_optuna.py`

**Input Data:**

- Parameter space definition: `configs/optuna_space.yaml`
- Analysis pipeline from previous sections

**Processing Content:**

2. Automatic parameter search using Optuna (100 trials)
3. Evaluation of known ruin reproduction rate for each trial
4. Determination and saving of optimal parameters

**Output Data:**
- `data/output/optuna/best_params.json` - Optimal parameters
- `optuna.db` - Optimization history database

### ⚙️ Optimization Strategy
